```python
import random

bases = ['A', 'G', 'T', 'C']
```

```python

genome_len = 10
class indiv: # genome, fitness, constructor, print, ...
    # def __init__(self, genome_len): # constructor 
    def __init__(self): # constructor 
        self.fitness = 0
        self.genome = []
        for i in range(genome_len):
            self.genome.append(random.choice(bases))

    def calcFitness(self):
        self.fitness = 0
        for i in range(0,(len(self.genome))):
            if(self.genome[i] == 'A'): # count A's
                self.fitness +=1
    
    def mutate(self):
        self.genome[random.randint(0, len(self.genome)-1)] = random.choice(bases)

    def crossover(self, other):
        crossPoint = random.randint(0, len(self.genome) -1)
        for i in range(crossPoint, len(self.genome)):
            temp = self.genome[i]
            self.genome[i] =  other.genome[i]
            other.genome[i] = temp
                
    def __str__(self):
        return_str = "Fitness = " + str(self.fitness)
        return_str += "\nGenome: " + str(self.genome)
        return return_str
        # return f"Fitness: {self.fitness} \nGenome: {self.genome}"

    def print(self):
        print("Fitness: ",self.fitness)
        print("Genome ", self.genome)
    
    
    def copy(self, source):
        for i in range(len(source.genome)):
            self.genome[i] = source.genome[i]
```

```python
popSize = 10

class pop: 
    def __init__(self):
        self.population = []
        for i in range(popSize):
            self.population.append(indiv())
        self.bestFit = 0
        self.best = 0
        self.avgFit = 0
        self.calcStats()
    
    def generation(self):
        tempPop = pop()
        for i in range(0, popSize, 2):
            p1 = self.tourn()
            p2 = self.tourn()

            tempPop.population[i].copy(self.population[p1])
            tempPop.population[i+1].copy(self.population[p2])
            tempPop.population[i+1].crossover(tempPop.population[i+1])
            tempPop.population[i].mutate()
            tempPop.population[i+1].mutate()
        
        for i in range(0,popSize):
            self.population[i].copy(tempPop.population[i])

    def tourn(self): 
        best = random.randint(0,popSize-1)
        bestFit = self.population[best].fitness
        for i in range(5):
            p2 = random.randint(0, popSize-1)
            if(self.population[p2].fitness >  bestFit):
                bestFit = self.population[p2].fitness
                best =  p2
        return best

    def calcStats(self):
        self.avgFit = 0
        self.population[0].calcFitness()
        self.bestFit = self.population[0].fitness
        self.best = 0
        for i in range(len(self.population)):
            self.population[i].calcFitness() # update fitness
            if(self.population[i].fitness > self.bestFit): #compare fitness
                self.bestFit = self.population[i].fitness
                self.best  = i
            self.avgFit += self.population[i].fitness
        self.avgFit = self.avgFit/len(self.population)

    
```

```python
p = pop()
p.calcStats()
for g in range (0, 10):
    print("Generation: ", g)
    print(p.population[p.best])
    print("Avg fit = ",p.avgFit)
    p.generation()
    p.calcStats()
    
print("Generation: ", g)
print(p.population[p.best])
print("Avg fit = ",p.avgFit)
```
