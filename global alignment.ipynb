```python
str1 = "HEAGAWGHEE"
str2 = "PAWHEAE"
gapP = -8

def my_print(array):
    for r in array:
        for c in r:
            print(f"{c:4}",end = ' ')
        print()

score = [[0 for x in range(0,len(str1)+1)] for y in range(0,len(str2)+1)]
direction = [[0 for x in range(0,len(str1)+1)] for y in range(0,len(str2)+1)]
for c in range(0,len(str1)+1):
    score[0][c] = c*gapP
    direction[0][c] = '-'
for r in range(0,len(str2)+1):
    score[r][0] = r*gapP
    direction[r][0] = '|'
direction[0][0] = 'X'
my_print(score)
my_print(direction)

def NW():
    for c in range(1,len(str1)+1):
        for r in range(1,len(str2)+1):
            scoreD = score[r-1][c-1] + blosum50_matrix[amino_acids[str2[r-1]]][amino_acids[str1[c-1]]]
            scoreV = score[r-1][c] + gapP
            scoreH = score[r][c-1] + gapP
            score[r][c] = max(scoreD,scoreV,scoreH)
            if(score[r][c] == scoreD):
                direction[r][c] = "\\"
            if(score[r][c] == scoreH):
                direction[r][c] = "-"
            if(score[r][c] == scoreV):
                direction[r][c] = "|"

def alignment():
    align1 = ""
    align2 = ""
    row = len(str2)
    col = len(str1)
    while(row > 0 or col > 0):
        if(direction[row][col] == "\\"):
            align1 += str1[col-1]
            align2 += str2[row-1]
            row -= 1
            col -= 1
        if(direction[row][col] == "-"):
            align1 += str1[col-1]
            align2 += "-"
            col -= 1
        if(direction[row][col] == "|"):
            align1 += "-"
            align2 += str2[row-1]
            row -= 1
    align1 = align1[::-1]
    align2 = align2[::-1]
    return align1,align2
```

```python
NW()
my_print(score)
print()
my_print(direction)
s1, s2 = alignment()
print(s1)
print(s2)
print(str1)
print(str2)
```
