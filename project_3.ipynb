```python
from collections import Counter
import math

count =  100
counting = 1
data1_list = []
unique_list = []
combined = {}
seq_length = 0

with open("DataFile1.txt", "r") as file:
    while count != 0:
        line = file.readline()

        # Set returns a set of uniquie characters in the string
        uni_list = set(line.strip())

        #Tyrns the unique set of characters to to a dict with the vaalue 0
        uni_list_int = dict.fromkeys(uni_list, 0)

        # runs througgh  all the unique keys and counts the number times iits in the string
        # adds it to the dict
        for i in uni_list_int.keys():
            uni_list_int[i] = line.count(i)

        # Alphabeticallu orders te string
        uni_list_int = dict(sorted(uni_list_int.items()))

        # once it goes through the first string for every other string 
        if counting >= 1:

            # the counter function from the collection class can
            # add the two dictionaries together based on keys and add the values to a single dict
            final_unique = dict(Counter(uni_list_int) + Counter(combined))

        # Alphabitize the combined dict
        final_unique = dict(sorted(final_unique.items()))

        # ## prints sorted dictionary
        # print(dict(sorted(uni_list_int.items())))

        ## checks if new character is in alrerad known list then if not added to listy

        data1_list.append(line.strip())
        seq_length += len(line.strip())

        if counting <= 1:
            ##  stores the first string onm the forst loopp to latyer use when combining the dicts
            combined = uni_list_int.copy()
        else:
            # stores final combined dicts from all the strings
            combined = final_unique.copy()
        counting += 1
        count -= 1

        combined = dict(sorted(combined.items()))

print(data1_list)
print(dict(sorted(combined.items())))
print("Seq Lenght: ", seq_length)
```

```python

# Calculates the frequenmcy of each letter over the seq_length
freq_letters  = {}

for key, value in combined.items():
    freq_letters[key] = value/ seq_length
    
print("Total: ", freq_letters)
```

```python
 #Calculate sthe frequncy that each pair occurs over all the strings
pair_counts = {}

# Compare each alignment with every other alignment
for align1 in range(len(data1_list)):
    for align2 in range(len(data1_list)):
        # Skip comparing an alignment with itself
        if align1 == align2:
            continue
        
        # Compare characters at each position in the two alignments
        for i in range(min(len(data1_list[align1]), len(data1_list[align2]))):
            pair = (data1_list[align1][i], data1_list[align2][i])
            # opposite_pair = (data1_list[align2][i], data1_list[align1][i])
            
            if pair in pair_counts:
                pair_counts[pair] += 1
            # elif pair[::-1] in pair_counts: # if you want to remove the reducancy in pairs l;ik a c or c a
            #     pair_counts[pair[::-1]] += 1
            else:
                pair_counts[pair] = 1

# Sort the dictionary by keys
pair_counts = dict(sorted(pair_counts.items()))

pair_counts
```

```python
# calculattes all the pairs that occured by adding all the pairs together
total_lenght = 0

for key, value in pair_counts.items():
    total_lenght += value

# calculate the scale by dividiing the pair count by the total length of all the pairs
paired_scaled = {}

for key, value in pair_counts.items():
    paired_scaled[key] = value / total_lenght

paired_scaled
```

```python
score = {}

scaling_factor = 0.25

for val1, val2 in pair_counts.keys():
    score[val1, val2] = math.ceil((1/scaling_factor) * math.log2(paired_scaled[val1, val2] / (freq_letters[val1] * freq_letters[val2])))
    score[val2, val1] = math.ceil((1/scaling_factor) * math.log2(paired_scaled[val1, val2] / (freq_letters[val2] * freq_letters[val1])))

score
```

```python
import pandas as pd

# Get all unique keys from rows and columns
residues = sorted(set(combined))

# Create a DataFrame initialized with zeros or NaNs
df = pd.DataFrame(index=residues, columns=residues, data=0.0)

# Fill in the values
for (i, j), val in score.items():
    df.at[i, j] = val


# df.to_csv("100_seq_score.csv")

df
```
